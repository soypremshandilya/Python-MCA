```python
import numpy as np 
import matplotlib.pyplot as plt

PI = 3.14
start = 1
end = 3600
stepsize = 1
X = np.arange(start, end, stepsize)

omega = 1 
omega2  = 3

Y = np.sin(2* PI * omega *X)
Y2 = 2* np.sin(2* PI* omega2* X)

plt.figure()
plt.plot(X,Y, label="label1", color = "Orange")

plt.plot(X, Y2, label= "label 2", color ="Black")
plt.legend()
plt.show()


```
